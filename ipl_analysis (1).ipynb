```python
#loading the required libraries
import pandas as pd
from  matplotlib import pyplot  as plt
import seaborn as sns
```

```python
#loading the ipl matches daataset
ipl=pd.read_csv('matches.csv')
```

```python
#having a glance at the first five records of dataset
ipl.head()
```

```python
#looking at the number of rows and colums in dataset
ipl.shape
```

```python
#getting the frequency of most man of the match award
ipl['player_of_match'].value_counts()
```

```python
#getting the top 10 players with most man of the match awards
ipl['player_of_match'].value_counts()[0:10]
```

```python
#getting the top 5 player with most man of the match awards
ipl['player_of_match'].value_counts()[0:5]
```

```python
list(ipl['player_of_match'].value_counts().keys())
```

```python
#making a bar_plot for the top 5 players with most man of the match awards
plt.figure(figsize=(8,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),list(ipl['player_of_match'].value_counts()[0:5]),color="r")
plt.show()
```

```python
#getting the frequency of result coloumn
ipl['result'].value_counts()
```

```python
#finding out the number of toss wins  w.r.t each team
ipl['toss_winner'].value_counts()
```

```python
#Extracting the records where a team won batting first 
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
#looking at the head
batting_first.head()
```

```python
#making a histogram
plt.figure(figsize=(7,7))
plt.hist(batting_first['win_by_runs'])
plt.show()
```

```python
#finding out the numbers of wins w.r.t each team after batting first 
batting_first['winner'].value_counts()
```

```python
#making a bar plot of top three teams with most wins after batting first 
plt.figure(figsize=(7,7))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]),color=["blue","orange","yellow"])
plt.show()
```

```python
#making a piechart 
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show
```

```python
#Extracting those records where a team has won after batting  second
batting_second=ipl[ipl['win_by_wickets']!=0]
batting_second
```

```python
#looking at the head
batting_second.head()
```

```python
#making a historgram for frequency of w.r.t number of wickets 

plt.figure(figsize=(7,7))
plt.hist(batting_second['win_by_wickets'],bins=40)
plt.show()
```

```python
#finding out the frequency number of wins w.r.t each time after batting second 
batting_second['winner'].value_counts()
 
```

```python
#making a bar plot for top -3  teams with most wins after batting second 
plt.figure ( figsize=(5,5))
plt.bar ( list(batting_second['winner'].value_counts()[0:3].keys()),list(batting_second['winner'].value_counts()[0:3]),color=["black","yellow","hotpink"])
plt.show ()
```

```python
#making a pie chart for distribution of most wins after batting second 
plt.figure(figsize=(7,7))
plt.pie( list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
